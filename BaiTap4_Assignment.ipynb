```python
import pandas as pd

#Đọc dữ liệu từ các file *.csv bằng hàm read_csv() của pandas
train_df = pd.read_csv('/content/PythonAssignment/train.csv')
#In 5 dòng dữ liệu đầu tiên
train_df.head()
```

```python
#In đầy đủ sơ lược dataframe
train_df.info()
```

##Chuẩn hóa dữ liệu

```python
#Trả về series gồm những dòng dữ liệu, những dòng dữ liệu trùng lặp được sẽ được mark là True
train_df.duplicated(keep=False)
```

```python
#Xóa những dòng dữ liệu trùng lặp
train_df.drop_duplicates()
```

```python
#Tính số lượng cell có giá trị Null ở mỗi cột
train_df.isnull().sum()
```

```python
#Kết quả phía trên cho thấy cột 'Postal Code' có 11 cell có giá trị NULL
#Thay thế các giá trị NULL ở cột 'Postal Code' bằng một giá trị khác
train_df['Postal Code'] = train_df['Postal Code'].fillna('43011.0')
```

```python
#Kiểm tra số lượng cell có giá trị Null ở mỗi cột sau khi chuẩn hóa
train_df.isnull().sum()
```

```python
#Format những cột có kiểu dữ liệu datetime về format Y/m/d
train_df['Order Date'] = pd.to_datetime(train_df['Order Date'],format='%d/%m/%Y')
train_df['Ship Date'] = pd.to_datetime(train_df['Ship Date'],format='%d/%m/%Y')
```

```python
train_df.head()
```

```python
#Xắp xếp các dòng dữ liệu theo cột 'Order Date' tăng dần
train_df.sort_values(by=['Order Date'],inplace=True)
#In 3 dòng dữ liệu cuối cùng
train_df.tail()
```

##Biểu đồ doanh số biến động theo tháng

```python
 import datetime
  
sales = train_df
#Tạo mới cột 'Order Month', lấy 'Y-M' từ cột 'Order Date' đổ vào cột mới này
sales['Order Month'] = sales['Order Date'].apply(lambda x: x if type(x) == str else x.strftime('%Y-%m'))
sales.head()
```

```python
#Nhóm dataframe theo cột 'Order Month' và tính tổng trên cột 'Sales'
sales = sales.groupby('Order Month')['Sales'].sum()
sales
```

```python
import matplotlib.pyplot as plt
#Vẽ biểu đồ biến động doanh số hàng tháng
plt.figure(num=None, figsize=(17, 4), dpi=80, facecolor='w', edgecolor='k')
sales.plot(xlabel='Tháng', ylabel='Doanh số')
plt.title('Doanh số biến động theo tháng')
plt.show()
```
