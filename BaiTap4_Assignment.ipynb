##Vẽ biểu đồ thể hiện biến động doanh số theo tháng. Cho biết doanh thu trung bình trong 3 tháng gần nhất.

###Biểu đồ thể hiện biến động doanh số theo tháng

```python
 
sales = train_df
#Tạo mới cột 'Order Month', lấy 'Y-M' từ cột 'Order Date' đổ vào cột mới này
sales['Order Month'] = sales['Order Date'].apply(lambda x: x if type(x) == str else x.strftime('%Y-%m'))
#Nhóm dataframe theo cột 'Order Month' và tính tổng trên cột 'Sales'
sales = sales.groupby('Order Month')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
#Vẽ biểu đồ biến động doanh số hàng tháng
plt.figure(num=None, figsize=(16, 5), dpi=80, facecolor='w', edgecolor='k')
sales.plot(xlabel='Năm-Tháng', ylabel='Doanh số')
plt.title('Biến động doanh số theo tháng')
plt.show()
```

###Doanh thu trung bình 3 tháng gần nhất

```python
#Lấy 3 elements cuối trong series sales
sales = train_df.groupby(pd.Grouper(key="Order Date", freq="3M"))['Sales'].mean()
```

```python
#Vẽ lưới biểu đồ
fig, ax = plt.subplots(num=None, sharey=True, figsize = (16,5))

#x lấy từ sales.index, sau đó format theo 'Y-m'
x = sales.index.to_series().apply(lambda a: a if type(a) == str else a.strftime('%Y-%m'))
x = x.values.astype(str)
#y lấy từ values của sales
y = sales.values

#Vẽ biểu đồ dạng bar
ax.bar(x, y)
ax.set_xlabel('Năm-Tháng')
ax.set_ylabel('Doanh thu trung bình')
ax.set_title('Doanh thu trung bình 3 tháng gần nhất', fontsize=15)
plt.show()
```

##Vẽ biểu đồ doanh số bán hàng theo quý và cho biết quý nào có tổng doanh số cao nhất?

```python
sales = train_df
#Nhóm datatframe theo cột 'Order Date', sau đó nhóm dữ liệu trong cột 'Order Date' theo quý, tính tổng ở cột 'Sales'
sales = sales.groupby(sales['Order Date'].dt.to_period('Q'))['Sales'].sum()
```

```python
#Vẽ lưới biểu đồ
fig, ax = plt.subplots(num=None, sharey=True, figsize = (16,5))

#Vẽ biểu đồ dạng bar
x = sales.index.astype(str)
y = sales.values

ax.bar(x, y)
ax.set_xlabel('Quý')
ax.set_ylabel('Doanh số')
ax.set_title('Doanh số theo quý', fontsize=15)
plt.show()
```

Theo biểu đồ biến động doanh số theo các quý như trên, ta thấy **Quý 4 năm 2018 có tổng doanh số cao nhất**.

##Tính thời gian trung giao hàng trung bình (ship date - order date) theo từng tháng và vẽ biểu đồ mô tả sự biến thiên của con số này

```python
sales = train_df
#Tạo mới cột 'Order Month', lấy 'Y-M' từ cột 'Order Date' đổ vào cột mới này
sales['Order Month'] = sales['Order Date'].apply(lambda x: x if type(x) == str else x.strftime('%Y-%m'))
#Tạo mới cột 'ShipAverage', giá trị cột 'ShipAverage' = (Ship Date - Order Date )
sales['Ship Average'] = sales['Ship Date'] - sales['Order Date']

#Nhóm theo cột 'Order Month', và tính trung bình ở cột 'Ship Average'
sales = sales.groupby('Order Month')['Ship Average'].mean()
```

```python
#Vẽ biểu đồ biến động thời gian ship trung bình
plt.figure(num=None, figsize=(16, 5), dpi=80, facecolor='w', edgecolor='k')
sales.plot(xlabel='Năm-Tháng', ylabel='Trung bình thời gian ship')
plt.title('Biến động trung bình thời gian ship')
plt.show()
```

##Cho biết các bang dẫn đầu doanh số theo từng vùng (east, west, south, central,..)

```python
# Tạo các danh sách, mỗi danh sách là một vùng của United State và gồm các tiểu bang trong vùng
northEast = ['Maine', 'Vermont', 'New Hampshire', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'New Jersey', 'Pennsylvania']
midwest = ['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin', 'Iowa', 'Kansas', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'South Dakota']
south = ['Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'South Carolina', 'Virginia', 'West Virginia', 'District of Columbia', 'Alabama', 'Kentucky', 'Mississippi', 'Tennessee', 'Arkansas', 'Louisiana', 'Oklahoma', 'Texas']
west = ['Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Utah', 'Wyoming', 'Alaska', 'California', 'Hawaii', 'Oregon', 'Washington']
```

```python
#Nhóm theo bang, tính tổng trên cột 'Sales'
sales_state = train_df.groupby(['State'])['Sales'].sum()
#Convert series sang dataframe và reset index
sales_state = sales_state.to_frame().reset_index()

#Tạo list chứa các vùng tương ứng của tiểu bang
areas = []

for i in range(len(sales_state)):
   x = 'northEast' if sales_state['State'].iloc[i] in northEast else ('midwest' if sales_state['State'].iloc[i] in midwest else ('south' if sales_state['State'].iloc[i] in south else ('west' if sales_state['State'].iloc[i] in west else 'NaN')))
   areas.append(x)

sales_state['Area'] = areas

#Nhóm theo các cột ['Area','State'], tính tổng trên cột 'Sales'
sales = sales_state.groupby(['Area','State'])['Sales'].sum()
#Tao series cho từng vùng
salesMidwest = sales.loc['midwest']
salesNorthEast = sales.loc['northEast']
salesSouth = sales.loc['south']
salesWest = sales.loc['west']
```

```python
import numpy as np
#Vẽ lưới biểu đồ
fig, ax = plt.subplots(2,2, figsize=(18,14))

#Vẽ biểu đồ dạng bar
N=9
ind = np.arange(N)
ax[0,0].bar(salesNorthEast.index.astype(str), salesNorthEast.values)
ax[0,0].set_title('North East')
ax[0,0].set_xticklabels(labels=salesNorthEast.index.astype(str),rotation=45, ha='right')

ax[0,1].bar(salesMidwest.index.astype(str), salesMidwest.values)
ax[0,1].set_title('Mid-West')
ax[0,1].set_xticklabels(salesMidwest.index.astype(str),rotation=45, ha='right')


ax[1,0].bar(salesSouth.index.astype(str), salesSouth.values)
ax[1,0].set_title('South')
ax[1,0].set_xticklabels(salesSouth.index.astype(str),rotation=45, ha='right')

ax[1,1].bar(salesWest.index.astype(str), salesWest.values)
ax[1,1].set_title('West')
ax[1,1].set_xticklabels(salesWest.index.astype(str),rotation=45, ha='right')

plt.suptitle('Doanh số theo từng vùng', fontsize=15)
plt.show()
```

Xem qua biểu đồ 'Doanh số theo từng vùng', ta thấy:
* North East: bang **New York** có doanh số dẫn đầu vùng này.
* Mid-West: bang **Illinois** có doanh số dẫn đầu vùng này.
* South: bang **Texas** có doanh số dẫn đầu vùng này.
* West: bang **California** có doanh số dẫn đầu vùng này.
